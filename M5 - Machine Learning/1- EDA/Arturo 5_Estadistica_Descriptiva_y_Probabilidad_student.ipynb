## Probabilidad



La Regla de Laplace:  $P(A) = \frac {\text{Nº casos favorables}}{\text{Nº casos totales}}$


- Si lanzamos un dado 🎲 tiene 6 lados posibles: {1,2,3,4,5,6}.
- Cada lado puede salir con la misma probabilidad... siempre que el dado no esté trucado 😬

Cuando lanzamos un dado, sabemos que la probabilidad de que salga cualquier número es $P(A) = \frac {1}{6}$ siendo `A = Probabilidad de que salga un 6`

## Distribución Gaussiana

El siguiente ejemplo ha sido extraído de la librería de numpy [aquí](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html).

```python
import numpy as np

mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

Comprobamos que la media de los datos generados da cero, el valor guardado en 'mu'.

```python
abs(mu - np.mean(s))
```

y la desviación también es la deseada.

```python
abs(sigma - np.std(s, ddof=1))
```

A continuación vamos a dibujar el histograma de las muestras generadas junto con la función de distribución de probabilidad para ver que dichas muestras siguen dicha distribución.

Ten cuenta que la función de densidad de probabilidad de una Gaussiana es:

$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{{(x-\mu)}^2}{2\sigma^2}}$

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
