## Distribuci칩n Gaussiana

El siguiente ejemplo ha sido extra칤do de la librer칤a de numpy [aqu칤](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html).

```python
import numpy as np

mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

Comprobamos que la media de los datos generados da cero, el valor guardado en 'mu'.

```python
abs(mu - np.mean(s))
```

y la desviaci칩n tambi칠n es la deseada.

```python
abs(sigma - np.std(s, ddof=1))
```

A continuaci칩n vamos a dibujar el histograma de las muestras generadas junto con la funci칩n de distribuci칩n de probabilidad para ver que dichas muestras siguen dicha distribuci칩n.

Ten cuenta que la funci칩n de densidad de probabilidad de una Gaussiana es:

$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{{(x-\mu)}^2}{2\sigma^2}}$

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
